## Load data

```python
dataRM = pd.read_csv('CSV/Oct5_00_RM.csv')
dataCP = pd.read_csv('CSV/Oct5_00_CP.csv')
dataAWEC = pd.read_csv('CSV/Oct5_00_AWEC.csv')
```

-----------------
## Apply Fitter Package
-----------------

```python
# fitter package with specific modifications for uwind @ RM
f_uRM = Fitter(dataRM['uwind'],
           distributions= ['weibull_min','rayleigh','gamma','genextreme','lognorm'],
           timeout=100,bins=30,xmin= -5, xmax = 10)

# fit the data
f_uRM.fit()

# return table and plot
f_uRM.summary()
```

```python
# fitter package with modifications for speed @ RM
f_sRM = Fitter(dataRM['speed'],
           distributions= ['weibull_min','rayleigh','gamma','genextreme','lognorm'],
           timeout=100,bins=30,xmin= 0, xmax = 10)
f_sRM.fit()
f_sRM.summary()
```

```python
# fitter package with modifications for gust @ RM
f_gRM = Fitter(dataRM['gust'],
           distributions= ['weibull_min','rayleigh','gamma','genextreme','lognorm'],
           timeout=100,bins=30,xmin= 3, xmax = 13)
f_gRM.fit()
f_gRM.summary()
```

```python
# fitter package with modifications for uwind @ CP
f_uCP = Fitter(dataCP['uwind'],
           distributions= ['weibull_min','rayleigh','gamma','genextreme','lognorm'],
           timeout=100,bins=30,xmin= -5, xmax = 10)
f_uCP.fit()
f_uCP.summary()
```

```python
# fitter package with modifications for speed @ CP
f_sCP = Fitter(dataCP['speed'],
           distributions= ['weibull_min','rayleigh','gamma','genextreme','lognorm'],
           timeout=100,bins=30,xmin= 0, xmax = 10)
f_sCP.fit()
f_sCP.summary()
```

```python
# fitter package with modifications for gust @ CP
f_gCP = Fitter(dataCP['gust'],
           distributions= ['weibull_min','rayleigh','gamma','genextreme','lognorm'],
           timeout=100,bins=30,xmin= 3, xmax = 13)
f_gCP.fit()
f_gCP.summary()
```

```python
# fitter package with modifications for uwind @ AWEC
f_uAWEC = Fitter(dataAWEC['uwind'],
           distributions= ['weibull_min','rayleigh','gamma','genextreme','lognorm'],
           timeout=100,bins=30,xmin= -5, xmax = 10)
f_uAWEC.fit()
f_uAWEC.summary()
```

```python
# fitter package with modifications for speed @ CP
f_sAWEC = Fitter(dataAWEC['speed'],
           distributions= ['weibull_min','rayleigh','gamma','genextreme','lognorm'],
           timeout=100,bins=30,xmin= 0, xmax = 10)
f_sAWEC.fit()
f_sAWEC.summary()
```

```python
# fitter package with modifications for gust @ CP
f_gAWEC = Fitter(dataAWEC['gust'],
           distributions= ['weibull_min','rayleigh','gamma','genextreme','lognorm'],
           timeout=100,bins=30,xmin= 3, xmax = 13)
f_gAWEC.fit()
f_gAWEC.summary()
```

## Figure

```python
# define xlim for each location and variable

# u-wind
x_u_RM = np.linspace(2,9,30)
x_u_CP = np.linspace(-3,0,30)
x_u_AWEC = np.linspace(-5,1,30)

# speed
x_s_RM = np.linspace(4,9,30)
x_s_CP = np.linspace(1,4,30)
x_s_AWEC = np.linspace(0,7,30)

# gust
x_g_RM = np.linspace(7,13,30)
x_g_CP = np.linspace(3,6,30)
x_g_AWEC = np.linspace(3,7,30)

# Rocky Mountain u-wind data and best fit pdf
RM_u = dataRM['uwind']
RM_u_pdf = f_uRM.fitted_pdf['weibull_min']

# Colorado Plateau u-wind data and best fit pdf
CP_u = dataCP['uwind']
CP_u_pdf = f_uCP.fitted_pdf['genextreme']

# Alta Wind Energy Center u-wind data and best fit pdf
AWEC_u = dataAWEC['uwind']
AWEC_u_pdf = f_uAWEC.fitted_pdf['genextreme']

# Rocky Mountain speed
RM_s = dataRM['speed']
RM_s_pdf = f_sRM.fitted_pdf['lognorm']

# Colorado Plateau speed
CP_s = dataCP['speed']
CP_s_pdf = f_sCP.fitted_pdf['weibull_min']

# Alta Wind Energy Center speed
AWEC_s = dataAWEC['speed']
AWEC_s_pdf = f_sAWEC.fitted_pdf['lognorm']

# Rocky Mountain gust
RM_g = dataRM['gust']
RM_g_pdf = f_gRM.fitted_pdf['lognorm']

# Colorado Plateau gust
CP_g = dataCP['gust']
CP_g_pdf = f_gCP.fitted_pdf['gamma']

# Alta Wind Energy Center gust
AWEC_g = dataAWEC['gust']
AWEC_g_pdf = f_gAWEC.fitted_pdf['lognorm']
```

```python
# define subplots
fig, ax = plt.subplots(3, 3)

# Rocky Mountain u-wind
ax[0, 0].hist(RM_u, density = True, color = 'k', edgecolor = 'w')
ax[0, 0].plot(x_u_RM, RM_u_pdf, color = 'b', label = 'weibull')
ax[0, 0].set_title("RM", size = 17)
ax[0, 0].set_xlabel("U-wind (m/s)", size = 14)
ax[0, 0].set_ylabel("Frequency", size = 15, labelpad = 10)
ax[0, 0].xaxis.set_tick_params(labelsize=13)
ax[0, 0].yaxis.set_tick_params(labelsize=13)
ax[0, 0].set_yticks([0,0.2,0.4])
ax[0, 0].legend(fontsize = 13)

# Colorado Plateau u-wind
ax[0, 1].hist(CP_u, density = True, color = 'k', edgecolor = 'w')
ax[0, 1].plot(x_u_CP, CP_u_pdf, color = 'forestgreen', label = 'GEV')
ax[0, 1].set_title("CP", size = 17)
ax[0, 1].set_xlabel("U-wind (m/s)", size = 14)
ax[0, 1].xaxis.set_tick_params(labelsize=13)
ax[0, 1].yaxis.set_tick_params(labelsize=13)
ax[0, 1].set_yticks([0,0.4,0.8])
ax[0, 1].legend(fontsize = 13)

# Alta Wind Energy Center u-wind
ax[0, 2].hist(AWEC_u, density = True, color = 'k', edgecolor = 'w')
ax[0, 2].plot(x_u_AWEC, AWEC_u_pdf, color = 'r', label = 'GEV')
ax[0, 2].set_title("AWEC", size = 17)
ax[0, 2].set_xlabel("U-wind (m/s)", size = 14)
ax[0, 2].xaxis.set_tick_params(labelsize=13)
ax[0, 2].yaxis.set_tick_params(labelsize=13)
ax[0 ,2].set_yticks([0,0.2,0.4])
ax[0, 2].legend(fontsize = 13)

# Rocky Mountain speed
ax[1, 0].hist(RM_s, density = True, color = 'k', edgecolor = 'w')
ax[1, 0].plot(x_s_RM, RM_s_pdf, color = 'b', label = 'lognorm')
ax[1, 0].set_xlabel("Speed (m/s)", size = 14)
ax[1, 0].set_ylabel("Frequency", size = 15, labelpad = 10)
ax[1, 0].xaxis.set_tick_params(labelsize=13)
ax[1, 0].yaxis.set_tick_params(labelsize=13)
ax[1 ,0].set_yticks([0,0.2,0.4])
ax[1, 0].legend(fontsize = 12.5)

# Colorado Plateau speed
ax[1, 1].hist(CP_s, density = True, color = 'k', edgecolor = 'w')
ax[1, 1].plot(x_s_CP, CP_s_pdf, color = 'forestgreen', label = 'weibull')
ax[1, 1].set_xlabel("Speed (m/s)", size = 14)
ax[1, 1].xaxis.set_tick_params(labelsize=13)
ax[1, 1].yaxis.set_tick_params(labelsize=13)
ax[1 ,1].set_yticks([0,0.4,0.8])
ax[1, 1].legend(fontsize = 13)

# Alta Wind Energy Center speed
ax[1, 2].hist(AWEC_s, density = True, color = 'k', edgecolor = 'w')
ax[1, 2].plot(x_s_AWEC, AWEC_s_pdf, color = 'r', label = 'lognorm')
ax[1, 2].set_xlabel("Speed (m/s)", size = 14)
ax[1, 2].xaxis.set_tick_params(labelsize=13)
ax[1, 2].yaxis.set_tick_params(labelsize=13)
ax[1 ,2].set_yticks([0,0.2,0.4])
ax[1, 2].legend(fontsize = 13)

# Rocky Mountain gust
ax[2, 0].hist(RM_g, density = True, color = 'k', edgecolor = 'w')
ax[2, 0].plot(x_g_RM, RM_g_pdf, color = 'b', label = 'lognorm')
ax[2, 0].set_xlabel("Gust (m/s)", size = 14)
ax[2, 0].set_ylabel("Frequency", size = 15, labelpad = 10)
ax[2 ,0].xaxis.set_tick_params(labelsize=13)
ax[2 ,0].yaxis.set_tick_params(labelsize=13)
ax[2 ,0].set_yticks([0,0.2,0.4])
ax[2, 0].legend(fontsize = 12.5)

# Colorado Plateau gust
ax[2, 1].hist(CP_g, density = True, color = 'k', edgecolor = 'w')
ax[2, 1].plot(x_g_CP, CP_g_pdf, color = 'forestgreen', label = 'gamma')
ax[2, 1].set_xlabel("Gust (m/s)", size = 14)
ax[2, 1].xaxis.set_tick_params(labelsize=13)
ax[2, 1].yaxis.set_tick_params(labelsize=13)
ax[2 ,1].set_yticks([0,0.4,0.8])
ax[2, 1].legend(fontsize = 13)

# Alta Wind Energy Center gust
ax[2, 2].hist(AWEC_g, density = True, color = 'k', edgecolor = 'w')
ax[2, 2].plot(x_g_AWEC, AWEC_g_pdf, color = 'r', label = 'lognorm')
ax[2, 2].set_xlabel("Gust (m/s)", size = 14)
ax[2, 2].xaxis.set_tick_params(labelsize=13)
ax[2, 2].yaxis.set_tick_params(labelsize=13)
ax[2 ,2].set_yticks([0,0.3,0.6])
ax[2, 2].legend(fontsize = 13)
    
plt.tight_layout(2.2)
plt.show()
```
