```python
# Style Dictionary to standarize plotting
# scheme between different python scripts 
styledict = {'xtick.labelsize':21,
             'xtick.major.size':9,
             'xtick.major.width':1,
             'ytick.labelsize':21,
             'ytick.major.size':9,
             'ytick.major.width':1,
             'legend.framealpha':0.0,
             'legend.fontsize':15,
             'axes.labelsize':22,
             'axes.titlesize':22,
             'axes.linewidth':2,
             'figure.figsize':(16,12),
             'savefig.format':'jpg'}
plt.rcParams.update(**styledict)
```

```python
uwind = pd.read_csv('uwind_combined.csv', header = 0)
vwind = pd.read_csv('vwind_combined.csv', header = 0)
speed = pd.read_csv('Speed_combined.csv', header = 0)
gust = pd.read_csv('Gust.csv', header = 0)
```

```python
# visualize dataset
uwind
```

```python
# create df with all 90 distributions as a single column
uwind_ = pd.DataFrame(uwind.to_numpy().flatten(), columns = {"Distribution"})
vwind_ = pd.DataFrame(vwind.to_numpy().flatten(), columns = {"Distribution"})
speed_ = pd.DataFrame(speed.to_numpy().flatten(), columns = {"Distribution"})
gust_ = pd.DataFrame(gust.to_numpy().flatten(), columns = {"Distribution"})
```

```python
# plot u-wind histogram with top 3 labeled
uwind_['Distribution'].value_counts().plot(kind='bar', color = 'k')

plt.xticks(size = 15,rotation='vertical')
plt.title("Top Distributions for U-Wind", pad = 10)
plt.text(30,7,'Top 3 Distributions:', size = 25, color = 'b')
plt.text(32,6.65, "Gamma Class", size = 23, color = 'b')
plt.text(32,6.3, "Burr Class", size = 23, color = 'b')
plt.text(32,5.95, "Weibull Class", size = 23, color = 'b')
plt.savefig("uwind_histogram.jpg", bbox_inches="tight")

# show plot
plt.show()
```

```python
# plot v-wind histogram with top 3 labeled
vwind_['Distribution'].value_counts().plot(kind='bar', color = 'k')

plt.xticks(size = 15,rotation='vertical')
plt.title("Top Distributions for V-Wind", pad = 10)
plt.text(26,10.5,'Top 3 Distributions:', size = 25, color = 'b')
plt.text(28,10, "Burr Class", size = 23, color = 'b')
plt.text(28,9.5, "Johnsonsu", size = 23, color = 'b')
plt.text(28,9, "Gamma Class", size = 23, color = 'b')
plt.savefig("vwind_histogram.jpg", bbox_inches="tight")

# show plot
plt.show()
```

```python
# plot speed histogram with top 3 labeled
speed_['Distribution'].value_counts().plot(kind='bar', color = 'k')

plt.xticks(size = 15,rotation='vertical')
plt.title("Top Distributions for Speed", pad = 10)
plt.text(30,7,'Top 3 Distributions:', size = 25, color = 'b')
plt.text(32,6.65, "Gamma Class", size = 23, color = 'b')
plt.text(32,6.3, "Burr Class", size = 23, color = 'b')
plt.text(32,5.95, "Weibull Class", size = 23, color = 'b')
plt.savefig("speed_histogram.jpg", bbox_inches="tight")

# show plot
plt.show()
```

```python
# plot gust histogram with top 3 labeled
gust_['Distribution'].value_counts().plot(kind='bar', color = 'k')

plt.xticks(size = 15,rotation='vertical')
plt.title("Top Distributions for Gust", pad = 10)
plt.text(24,12.5,'Top 3 Distributions:', size = 25, color = 'b')
plt.text(26,11.9, "Burr Class", size = 23, color = 'b')
plt.text(26,11.3, "Weibull Class", size = 23, color = 'b')
plt.text(26,10.7, "laplace_asymmetric", size = 23, color = 'b')
plt.savefig("gust_histogram.jpg", bbox_inches="tight")

# show plot
plt.show()
```
