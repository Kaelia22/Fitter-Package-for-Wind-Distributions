```python
# Style Dictionary to standarize plotting
# scheme between different python scripts 
styledict = {'xtick.labelsize':21,
             'xtick.major.size':9,
             'xtick.major.width':1,
             'ytick.labelsize':21,
             'ytick.major.size':9,
             'ytick.major.width':1,
             'legend.framealpha':0.0,
             'legend.fontsize':15,
             'axes.labelsize':22,
             'axes.titlesize':22,
             'axes.linewidth':2,
             'figure.figsize':(16,12),
             'savefig.format':'jpg'}
plt.rcParams.update(**styledict)
```

```python
# read data
uwind = pd.read_csv('uwind_o.csv', header = None)
vwind = pd.read_csv('vwind_o.csv', header = None)
speed = pd.read_csv('Speed_o.csv', header = None)
gust = pd.read_csv('Gust_o.csv', header = None)
```

```python
# create df with all distributions in single column
uwind_ = pd.DataFrame(uwind.to_numpy().flatten(), columns = {"Distribution"})
vwind_ = pd.DataFrame(vwind.to_numpy().flatten(), columns = {"Distribution"})
speed_ = pd.DataFrame(speed.to_numpy().flatten(), columns = {"Distribution"})
gust_ = pd.DataFrame(gust.to_numpy().flatten(), columns = {"Distribution"})
```

```python
# plot histogram 
plt.figure(figsize=(6, 18)) 
counts = uwind_["Distribution"].value_counts()
plt.barh(counts.index, counts, height = 0.5, color = 'k')

plt.yticks(fontsize = 15)
plt.xticks(fontsize = 17)
plt.title("U-wind Distributions", pad = 10)
plt.savefig("uwind_o.jpg",bbox_inches="tight")

# show plot
plt.show()
```
